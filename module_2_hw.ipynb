```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
import math
pi = math.pi
```

Question 4

```python
# Parameters
r = np.linspace(1.5, 2.5, num = 100) # angstroms
k = 150 # kcal/(mol*angstroms^2)
r0 = 2.02 # angstroms
```

```python
# Potential energy and plot
U = (k/2)*(r - r0)**2 # kcal/mol
plt.plot(r, U)
plt.xlabel("Bond Length (angstroms)")
plt.ylabel("Potential Energy (kcal/mol)")
plt.title("Potential Energy vs. Bond Length")
```

```python
F = -k*(r - r0) # kcal/(mol*angstroms)
plt.plot(r, F)
plt.xlabel("Bond Length (angstroms)")
plt.ylabel("Force (kcal/(mol*angstroms))")
plt.title("Force vs. Bond Length")
```

```python
F1 = -k*(r - r0)*69.2786 # pN
plt.plot(r, F1)
plt.xlabel("Bond Length (angstroms)")
plt.ylabel("Force (pN)")
plt.title("Force vs. Bond Length")
```

Question 5

```python
# Parameters
phi = np.linspace(0, 2*pi, 100) # radians
A = 1.5 # kcal/mol
```

```python
# Potential energy
fig = plt.figure()
ax = fig.add_subplot(111)
U = A*np.cos(3*phi) # kcal/mol
plt.plot(phi, U)
plt.xlabel("Torsional Angle (\u03A6, radians)")
plt.ylabel("Torsional Potential (kcal/mol)")
plt.title("Torsional Potential vs. Torsional Angle")

# Label local energy maxima and minima
maxima = np.array([0, 2*pi/3, 4*pi/3, 6*pi/3])
minima = maxima + (pi/3)
minima = minima[0:len(minima) - 1] # remove extraneous point
Umax = A*np.cos(3*maxima)
Umin = A*np.cos(3*minima)
for xy in zip(maxima, Umax):
    ax.annotate(' (%.2f, %.2f) Maxima' % xy, xy = xy, textcoords = 'data', fontsize = 8)
for xy in zip(minima, Umin):
    ax.annotate('   (%.2f, %.2f) Minima' % xy, xy = xy, textcoords = 'data', fontsize = 8)
plt.scatter(maxima, Umax, color = "black")
plt.scatter(minima, Umin, color = "black")

# Sample points for force 
posF = np.array([pi/6, 5*pi/6])
negF = np.array([9*pi/6 + pi/3])
UposF = A*np.cos(3*posF)
UnegF = A*np.cos(3*negF)
ax.annotate("", xy = (posF[0], UposF[0]), xytext = (posF[0]-0.2, UposF[0]+0.6), arrowprops = dict(arrowstyle = "fancy"))
ax.annotate("", xy = (posF[1], UposF[1]), xytext = (posF[1]-0.2, UposF[1]+0.6), arrowprops = dict(arrowstyle = "fancy"))
ax.annotate("", xy = (negF[0], UnegF[0]), xytext = (negF[0]+0.2, UnegF[0]+0.6), arrowprops = dict(arrowstyle = "fancy"))
for xy in zip(posF, UposF):
    ax.annotate(' (%.2f, %.2f) \n Force in \n +\u03A6' % xy, xy = xy, textcoords = 'data', fontsize = 8)
for xy in zip(negF, UnegF):
    ax.annotate('   (%.2f, %.2f) \n    Force in \n     \u03A6' % xy, xy = xy, textcoords = 'data', fontsize = 8)
plt.scatter(posF, UposF, color = "black")
plt.scatter(negF, UnegF, color = "black")
plt.axvspan(0, pi/3, color = "green", alpha = 0.3)
plt.axvspan(pi/3, 2*pi/3, color = "red", alpha = 0.3)
plt.axvspan(2*pi/3, 3*pi/3, color = "green", alpha = 0.3)
plt.axvspan(3*pi/3, 4*pi/3, color = "red", alpha = 0.3)
plt.axvspan(4*pi/3, 5*pi/3, color = "green", alpha = 0.3)
plt.axvspan(5*pi/3, 6*pi/3, color = "red", alpha = 0.3)
```

Question 6

```python
# Parameters
phi = np.linspace(0, 2*pi, 100)
A = 1.5 # kcal/mol
B = 2.0 # kcal/mol
```

```python
# Potential Energy
U = A*np.cos(3*phi) + B*np.cos(phi)
plt.plot(phi, U)
plt.xlabel("Torsional Angle (\u03A6, radians)")
plt.ylabel("Torsional Potential (kcal/mol)")
plt.title("Torsional Potential vs. Torsional Angle")
```

```python
# Force
F = 3*A*np.sin(3*phi) + B*np.sin(phi)
plt.plot(phi, F)
plt.xlabel("Torsional Angle (\u03A6, radians)")
plt.ylabel("Torsional Force (kcal/mol")
plt.title("Torsional Potential vs. Torsional Force")
```

```python
# Parameters
epsilon = 0.134 # kcal/mol
sigma = 3.18 # angstroms
r = np.linspace(sigma/1.05, 5*sigma, 100)
```

```python
# Potential Energy
U = 4*epsilon*((sigma/r)**12 - (sigma/r)**6) # kcal/mol
plt.plot(r, U)
plt.xlabel("Bond Length (angstroms)")
plt.ylabel("Potential Energy (kcal/mol)")
plt.title("Potential Energy vs. Bond Length")

# Minimum energy
r_min = (2**(1/6))*sigma # angstroms
U_min = 4*epsilon*((sigma/r_min)**12 - (sigma/r_min)**6) # kcal/mol
plt.scatter(r_min, U_min, color = "red")
```

```python
# Force
F1 = (48*epsilon*(sigma**12))/(r**13)
F2 = (-24*epsilon*(sigma**6))/(r**7)
F = F1 + F2 # kcal/(mol*angstroms)
plt.plot(r, F)
plt.xlabel("Bond Length (angstroms)")
plt.ylabel("Force (kcal/(mol*angstroms))")
plt.title("Force vs. Bond Length")

# Minimum energy
r_min = (2**(1/6))*sigma # angstroms
F_min = (48*epsilon*(sigma**12))/(r_min**13) + (-24*epsilon*(sigma**6))/(r_min**7) # kcal/(mol*angstroms)
plt.scatter(r_min, F_min, color = "red")
```

Question 10

```python
# Parameters
epsilon = 0.134 # kcal/mol
sigma = 3.18 # angstroms
r = np.linspace(3, 8, 100) # angstroms
r_diagonal = np.sqrt(r**2 + 2**2) # angstroms
qC = 0.5 # a.u.
qO = -0.5 # a.u.
```

```python
# Parallel case
# Lennard-Jones
LJ_CC_parallel = 4*epsilon*((sigma/r)**12 - (sigma/r)**6)
LJ_OO_parallel = 4*epsilon*((sigma/r)**12 - (sigma/r)**6)
LJ_CO_parallel = 4*epsilon*((sigma/r_diagonal)**12 - (sigma/r_diagonal)**6)
U_LJ_parallel = LJ_CC_parallel + LJ_OO_parallel + 2*LJ_CO_parallel
plt.plot(r, U_LJ_parallel)
plt.xlabel("r (angstroms)")
plt.ylabel("LJ Potential Energy (kcal/mol)")
plt.title("LJ Potential Energy vs. r (parallel)")
```

```python
# Parallel case
# Coulomb 
Coul_CC_parallel = (332*qC*qC)/(r)
Coul_OO_parallel = (332*qO*qO)/(r)
Coul_CO_parallel = (332*qC*qO)/(r_diagonal)
U_Coul_parallel = Coul_CC_parallel + Coul_OO_parallel + 2*Coul_CO_parallel
plt.plot(r, U_Coul_parallel)
plt.xlabel("r (angstroms)")
plt.ylabel("Coulomb Potential Energy (kcal/mol)")
plt.title("Coulomb Potential Energy vs. r (parallel)")
```

```python
# Parallel case
# Total energy
U_total_parallel = U_LJ_parallel + U_Coul_parallel
plt.plot(r, U_total_parallel)
plt.xlabel("r (angstroms)")
plt.ylabel("Total Potential Energy (kcal/mol)")
plt.title("Total Potential Energy vs. r (parallel)")
```

```python
# Anti-parallel case
# Lennard-Jones
LJ_CC_antiparallel = 4*epsilon*((sigma/r_diagonal)**12 - (sigma/r_diagonal)**6)
LJ_OO_antiparallel = 4*epsilon*((sigma/r_diagonal)**12 - (sigma/r_diagonal)**6)
LJ_CO_antiparallel = 4*epsilon*((sigma/r)**12 - (sigma/r)**6)
U_LJ_antiparallel = LJ_CC_antiparallel + LJ_OO_antiparallel + 2*LJ_CO_antiparallel
plt.plot(r, U_LJ_antiparallel)
plt.xlabel("r (angstroms)")
plt.ylabel("LJ Potential Energy (kcal/mol)")
plt.title("LJ Potential Energy vs. r (anti-parallel)")
```

```python
# Anti-parallel case
# Coulomb 
Coul_CC_antiparallel = (332*qC*qC)/(r_diagonal)
Coul_OO_antiparallel = (332*qO*qO)/(r_diagonal)
Coul_CO_antiparallel = (332*qC*qO)/(r)
U_Coul_antiparallel = Coul_CC_antiparallel + Coul_OO_antiparallel + 2*Coul_CO_antiparallel
plt.plot(r, U_Coul_antiparallel)
plt.xlabel("r (angstroms)")
plt.ylabel("Coulomb Potential Energy (kcal/mol)")
plt.title("Coulomb Potential Energy vs. r (anti-parallel)")
```

```python
# Anti-parallel case
# Total energy
U_total_antiparallel = U_LJ_antiparallel + U_Coul_antiparallel
plt.plot(r, U_total_antiparallel)
plt.xlabel("r (angstroms)")
plt.ylabel("Total Potential Energy (kcal/mol)")
plt.title("Total Potential Energy vs. r (anti-parallel)")
```
