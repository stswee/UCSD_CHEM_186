```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
```

Question 1

```python
# Define angles
phi = np.linspace(0, 2*np.pi, 100) # radians

# Determine potential energy
U = np.cos(3*phi) + np.sin(phi) # kcal/mol

# Plot
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(phi, U)
plt.xlabel("Torsional Angle (\u03A6, radians)")
plt.ylabel("Torsional Potential (kcal/mol)")
plt.title("Torsional Potential vs. Torsional Angle")

# Label minima
m1 = 5.185 # global minimum
m2 = 3.254 # lower local minima
m3 = 0.985 # higher local minima
U1 = np.cos(3*m1) + np.sin(m1) # kcal/mol
U2 = np.cos(3*m2) + np.sin(m2) # kcal/mol
U3 = np.cos(3*m3) + np.sin(m3) # kcal/mol
minima = [m1, m2, m3]
Umin = [U1, U2, U3]
plt.scatter(minima, Umin, color = "black") # Plot minima
ax.annotate("MG", xy = (m1, U1), xytext = (m1-0.15, U1 + 0.1)) # global minima
ax.annotate("M2", xy = (m2, U2), xytext = (m2-0.15, U2 + 0.1)) # m2
ax.annotate("M3", xy = (m3, U3), xytext = (m3-0.15, U3 + 0.1)) # m3
```
